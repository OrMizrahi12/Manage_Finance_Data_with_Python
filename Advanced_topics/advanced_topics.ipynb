# Shifting Dates with pd.DateOffeset()

```python
import pandas as pd
```

```python
SP500 = pd.read_csv("../Data/SP500.csv", parse_dates= ["Date"], index_col= "Date")
```

```python
SP500.index
```

```python
#SP500.index + 2 
```

```python
pd.DateOffset(days = 2)
```

```python
type(pd.DateOffset(days = 2))
```

```python
SP500.index
```

```python
# adding a time offset of 2 days to each datetime value
# in the index of the SP500 DataFrame
SP500.index + pd.DateOffset(days = 2)
```

```python
# adding a time offset of 3 weeks to each datetime value
# in the index of the SP500 DataFrame
SP500.index + pd.DateOffset(weeks = 3)
```

```python
# adding a time offset of 3 months to each datetime value
# in the index of the SP500 DataFrame
SP500.index + pd.DateOffset(months = 3)
```

```python
# adding a time offset of 10 years to each datetime value
# in the index of the SP500 DataFrame
SP500.index + pd.DateOffset(years = 10)
```

```python
# adding more complicated time offset:
SP500.index + pd.DateOffset(years = 1, months = 2, days = 15, hours = 12, minutes = 30)
```

```python
# substract day 
SP500.index - pd.DateOffset(days = 1)
```
