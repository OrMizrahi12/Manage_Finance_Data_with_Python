# Normalizing Time Series to a Base Value (100)

> because the stocks start from different level (e.g: AAPL start from 90, MSFT start from 60, and so on..), for compare them, we need to normelize on the same base. for example, base of 100: that all the stocks will start from base of 100.  

```python
close.head()
```

```python
# first column, first row
close.iloc[0,0]
```

```python
# Example of how to normalize the AAPL stock to base of 100
# divide all AAPL price with the first price of AAPL, then multiply it by base. 
close.AAPL.div(close.iloc[0,0]).mul(100)

# its normlize with base of 100
```

```python
close.iloc[0]
```

```python
# Normalize all the instrument to base 100
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()

# Now, we can compare them better. 
```

# The shift() method

```python
close.head()
```

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
# shift the close price one step ahead
aapl.shift(periods = 1)
```

```python
# lag featre
aapl["lag1"] = aapl.shift(periods = 1)
```

```python
aapl.head()
```

```python
# substract the price with the shifted price
aapl.AAPL.sub(aapl.lag1)
```

```python
# Compute the different between the price and its lag1 (the next day price)
aapl["Diff"] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl.head()
```

```python
# precent change in stock return   
aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
# So let's create a new column and calculate the return precentege
aapl["pct_change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl.head()
```

# Measuring Stock Perfromance with MEAN Return and STD of Returns

```python
import numpy as np
```

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
aapl.pct_change().dropna()
```

```python
# Compute the simple return (the precentege change of stock price)
ret = aapl.pct_change().dropna()
ret.head()
```

```python
ret.info()
```

```python
# Plot the distribution of the return 
ret.plot(kind = "hist", figsize = (12 ,8), bins = 100)
plt.show()
```

```python
# Compute the daily mean return 
daily_mean_Return = ret.mean()
daily_mean_Return
```

```python
# compute the daily variance (daily risk)
var_daily_Returns = ret.var()
var_daily_Returns
```

```python
std_daily_Returns = np.sqrt(var_daily_Returns)
std_daily_Returns
```

```python
ret.std()
```

```python
# Annuale mean return 
ann_mean_Return = ret.mean() * 252
ann_mean_Return
```

```python
# annualize risk (variance)
ann_var_Returns = ret.var() * 252
ann_var_Returns
```

```python
# Annualize risk 
ann_std_Returns = np.sqrt(ann_var_Returns)
ann_std_Returns
```

```python
# annualize risk 
ret.std() * np.sqrt(252)
```

# Financial Time Series - Return and Risk

- Let's analys and compare the mean return and risk of each instrument

```python
import numpy as np
```

```python
# Plot the normalize stocks (with norm base of 100)
norm.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
# close price of the instruments
close.head()
```

```python
# simple return 
close.pct_change().dropna()
```

```python
# Compute the simple return precentege
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
# describe statistics: extract the mean and std of return
summary = ret.describe().T.loc[:, ["mean", "std"]]
summary
```

```python
# we can also make that like this:
ret.describe().loc[["mean","std"]].T
```

```python
# Convert the mean return & risk to annualize:
summary["mean"] = summary["mean"]*252 # annualize mean return
summary["std"] = summary["std"] * np.sqrt(252) # annualize risk 
```

```python
summary
```

```python
# Plot the annualize mean and annualize risk for each instrument:

summary.plot(kind = "scatter", x = "std", y = "mean", figsize = (15,12), s = 50, fontsize = 15)

for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "std"]+0.002, summary.loc[i, "mean"]+0.002), size = 15)

plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 20)
plt.show()

# We can see the pattern: the more return - the more risk.
# the less return - the less risk 
```
